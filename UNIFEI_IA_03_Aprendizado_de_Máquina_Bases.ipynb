# Supervisionado Classificativo

```python
flores = ds.load_iris()
iris = pd.DataFrame(data= np.c_[flores['data'], flores['target']],
                     columns= flores['feature_names'] + ['Species'])
iris['Species'] = pd.Categorical.from_codes(flores.target, flores.target_names)
iris.head()
```

```python
iris.describe()
```

# Não Supervisionado

```python
X, y = ds.make_blobs(n_samples=300,centers=3,n_features=2)
grupos = pd.DataFrame(X)
grupos.head(10)
```

```python
grupos.describe()
```

```python
grupos.info()
```

```python
import matplotlib.pyplot as plt

plt.figure()
colors = ['#4EACC5', '#FF9C34', '#4E9A06', 'm']

for k, col in enumerate(colors):
    cluster_data = y == k
    plt.scatter(X[cluster_data, 0], X[cluster_data, 1],
                c=col, marker='.', s=10)

plt.title("Grupos")
plt.xticks([])
plt.yticks([])
plt.show()
```
